```python
"""
Day 30 Activity: EDA Exercise
Tasks:
1) Load dataset
2) Run univariate, bivariate, and correlation analysis
3) Write 3 insights
"""

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# TODO: Load data from data/day30_eda.csv
#histogram
df = pd.read_csv("/kaggle/input/datasets/rozanaali/aryomah1/day30_eda.csv")
plt.figure(figsize=(6, 4))
df["income"].hist(bins=30, edgecolor="black")
plt.title("Income Distribution — Histogram")
plt.xlabel("Income"); plt.ylabel("Count")
plt.tight_layout(); plt.show()
#histogram with bin size

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
df["age"].hist(bins=10, ax=axes[0]); axes[0].set_title("10 bins")
df["age"].hist(bins=30, ax=axes[1]); axes[1].set_title("30 bins")
df["age"].hist(bins=60, ax=axes[2]); axes[2].set_title("60 bins")
plt.show

# TODO: Perform basic EDA and write insights
```

```python
df = pd.read_csv("/kaggle/input/datasets/rozanaali/aryomah1/day30_eda.csv")
#KDE
plt.figure(figsize=(6, 4))
sns.kdeplot(df["income"], fill=True)
plt.title("Income Distribution — KDE")
plt.xlabel("Income"); plt.ylabel("Density")
plt.tight_layout(); plt.show()
#hist+KDE
plt.figure(figsize=(6, 4))
df["spend"].hist(bins=10, alpha=0.5, density=True, label="Histogram")
sns.kdeplot(df["spend"], color="red", label="KDE")
plt.legend(); plt.title("spend — Histogram + KDE")
plt.show
```

```python
df = pd.read_csv("/kaggle/input/datasets/rozanaali/aryomah1/day30_eda.csv")
#normal vs skewed data
sns.kdeplot(df["age"], label="Age (approx normal)")
sns.kdeplot(df["income"], label="Income (right-skewed)")
plt.legend(); plt.title("Normal vs Skewed Example")
```

```python
df = pd.read_csv("/kaggle/input/datasets/rozanaali/aryomah1/day30_eda.csv")
#pandas and seaborn
df.boxplot(column="income"); plt.title("pandas boxplot")
plt.show()
sns.boxplot(x="age", y="income", data=df); plt.title("seaborn age")
#Boxplot Anatomy
plt.figure(figsize=(6, 4))
plt.boxplot(df["income"], vert=True, showfliers=True)
plt.title("income — Boxplot"); plt.ylabel("income")
#Boxplots vs Histograms
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
df["income"].hist(bins=20, ax=axes[0]); axes[0].set_title("Histogram")
sns.boxplot(y="income", data=df, ax=axes[1]); axes[1].set_title("Boxplot")
```

```python
df = pd.read_csv("/kaggle/input/datasets/rozanaali/aryomah1/day30_eda.csv")
#correlations
corr_matrix = df[["age","income","spend"]].corr()
print(corr_matrix)
#Correlation Heatmaps
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show
```
